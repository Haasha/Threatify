```python
import cv2
import matplotlib.pyplot as plt
import time
```

```python
camera = cv2.VideoCapture('rtsp://root:pass@201.108.22.78:554/mpeg4/media.amp')
print(camera)
```

```python
def ReadCamera():
    count = 0
    while True:
        (grabbed, frame) = camera.read()
        if grabbed == True:
            count = 0
            yield frame
```

```python
for frame in ReadCamera():
    plt.imshow(frame)
    break
```
