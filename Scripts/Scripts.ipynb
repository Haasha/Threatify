```python
def ReadCamera():
    camera = cv2.VideoCapture('http://202.245.13.81:80/cgi-bin/camera?resolution=640&amp;quality=1&amp;Language=0&amp;COUNTER')
    count = 0
    while True:
        (grabbed, frame) = camera.read()
        if grabbed == True:
            count = 0
            yield frame
        else:
            print("FRAME NOT FOUND")
```
